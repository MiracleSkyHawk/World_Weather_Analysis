```python
import pandas as pd
import numpy as np
from citipy import citipy
from datetime import datetime
from config import weather_api_key
import json
import requests
```

```python
lats = np.random.uniform(low = -90.0, high = 90.0, size = 2000)
lgns = np.random.uniform(low = -180.0, high = 180.0, size = 2000)
lat_lngs = zip(lats, lgns)
coordinates = list(lat_lngs)
```

```python
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

```python
city_data = []
base_url = "http://api.openweathermap.org/data/2.5/weather?units=metric&appid=" + weather_api_key + "&q="

for i, city_name in enumerate(cities):
    url = base_url + city_name.replace(" ", "+")
    try:
        weather = requests.get(url).json()
        city_lat = weather["coord"]["lat"]
        city_lng = weather["coord"]["lon"]
        city_max_temp = weather["main"]["temp_max"]
        city_humidity = weather["main"]["humidity"]
        city_clouds = weather["clouds"]["all"]
        city_wind = weather["wind"]["speed"]
        city_country = weather["sys"]["country"]
        city_description = weather["weather"][0]["description"]
        city_date = datetime.utcfromtimestamp(weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        city_data.append({"City":city_name.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "City Description": city_description,
                          "Date": city_date})
    except:
        print("City not found. Skipping...")
        pass
    
    
```

```python
city_data_df = pd.DataFrame(city_data)
```

```python
city_data_df = city_data_df[["City",'Country','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','City Description']]
city_data_df
```

```python
city_data_df.to_csv("Weather_Database/WeatherPy_Database.csv",index_label="City_ID")
```
